# Introduction 

<Vagus nerve stimulation (VNS)>
The vagus nerve is a nerve that runs along the neck. It controls the functions of the heart, lungs and digestive tract, as well as sends updated sensory information regarding the state of these organs to the brain. 

The selectivity of neurons in sensory cortex, including the primary auditory cortex (A1), can be modified by pairing neuromodulator (acetylcholine and norepinephrine) release with sensory stimulation. Vagus nerve stimulation (VNS) is a safe and effective way to trigger the release of neuromodulators with a high degree of temporal control. By repeated pairing tonal sounds (sensory stimulation) with VNS, for example, alters neurons' spectral tuning and induces specific and long-lasting plasticity in A1. Based on the capacity to drive specific changes in neural circuitry, VNS paired with experience has been successful in effectively ameliorating animal models of chronic tinnitus, stroke, and posttraumatic stress disorder. Although there are evidences showing VNS-tone pairing can promote A1 cortical plasticity, there is no data showing how VNS-tone pairing would affect behavioral performances, such as tone or sound recognition. Hence, the Stephen lab is interested in finding out whether VNS-tone pairing can increase the speed of learning to identify different tones.

<Behavior study>
After surgery, animals were trained to perform a tone-reward association task using classical conditiong method. In the behavioral task, two tones (either higher or lower frequency) and references (noise) were presented to animals. One out of the two tones was associated with a liquid reward while the other tone has no reward. In one training session, for example, we set the higher tone to associate with a reward. In the next session, we set the lower tone to have a reward. This is to train animals to identify which tone was associated with a reward within a training session. # Using the 'for loop' for our data - Hit Rates


That's great, next let's try using that 'for loop' knowledge at getting the hit rate for every iteration of 5. Make a list called HR_t1VNS where we can store the averages of every 5 trials to look at learning over time. 

# Exercise 9

```python
# Answer
HR_t1VNS=[]
for i in range(0, len(t1VNS_hitcount), 5):
    HR_t1VNS.append(sum(t1VNS_hitcount[i:i+5]) / 5)
HR_t1VNS
```



Does this look right? What might be wrong? What if the total number of calculations isn't divisible by 5? Let's see how many observations were present in our data sample.



# Exercise 10

```python
# Answer
print(len(t1VNS_hitcount))
```



We need to figure out some way to account for extra observations that aren't multiples of 5. How can we keep track of where we're at in the list, and run the final number against the correct divisor? What can we add to our 'for loop' to do this? Hint: You can print inside a 'for loop'!




# Exercise 11

```python
# Answer
HR_t1VNS=[]
for i in range(0, len(t1VNS_hitcount), 5):
    print("i is: {0}".format(i))
    divisor = len(t1VNS_hitcount[i:i+5])
    print("length of list: {0}".format(divisor))
    HR_t1VNS.append((sum(t1VNS_hitcount[i:i+5]) / divisor))
```

You know, maybe it will be better to just remove that last number. A hit rate of 1 / 1 being compared directly against hit rates that had 5 values will artificially inflate the hit rate (albeit by a very small amount). How can you delete the last number?

# Exercise 12

```python
# Answer
HR_t1VNS = HR_t1VNS[:-1]
HR_t1VNS
```



Now lets do this for the 3 other values we care about, the hit rate while the low frequency (unrewarded tone) is played under the condition of vagus nerve stimulation, and the hit rate of rewarded and unrewarded tones without vagus nerve stimulation. Name the lists as HR_t2VNS, HR_t1UN, and HR_t2UN. Don't forget to check how many total data points you have and to delete any non-multiples of 5!



# Exercise 13

```python
# Answer
HR_t2VNS=[]
for i in range(0, len(t2VNS_hitcount), 5):
    #print("i is: {0}".format(i))
    divisor = len(t2VNS_hitcount[i:i+5])
    #print("length of list: {0}".format(divisor))
    HR_t2VNS.append((sum(t2VNS_hitcount[i:i+5]) / divisor))
HR_t2VNS=HR_t2VNS[:-1]  
print(HR_t2VNS)

HR_t1UN=[]
for i in range(0, len(t1UN_hitcount), 5):
    #print("i is: {0}".format(i))
    divisor = len(t1UN_hitcount[i:i+5])
    #print("length of list: {0}".format(divisor))
    HR_t1UN.append((sum(t1UN_hitcount[i:i+5]) / divisor))
HR_t1UN=HR_t1UN[:-1]  
print(HR_t1UN)
    
HR_t2UN=[]
for i in range(0, len(t2UN_hitcount), 5):
    #print("i is: {0}".format(i))
    divisor = len(t2UN_hitcount[i:i+5])
    #print("length of list: {0}".format(divisor))
    HR_t2UN.append((sum(t2UN_hitcount[i:i+5]) / divisor))
HR_t2UN=HR_t2UN[:-1]  
print(HR_t2UN)

```

# List Comprehension - another way of doing this

# Learning Index

Now we're finally close to really seeing how these compare to each other. We're going to look at this in terms of a "Learning Index". The Learning Index is defined as (Hit Rate of Tone 1 / Hit Rate of Tone 2) / (0.5 * (Hit Rate of Tone 1 + Hit Rate of Tone 2)). Since we're interested in how this looks over the course of the trials, we want to look at each compared block of 5 (already performed in our hit rate analysis). How can we do this? Hint: More For Loops! 

A value of 2 is a perfect score. They always lick during the correct tone, and never lick during the wrong tone. A score of -2 is the opposite, indicative of an unlearned ferret. A value of 0 indicates ferrets licking or not licking during both tones.  

# Exercise 16

```python
# Answer
LI_VNS=[]
for n in range(0, len(HR_t1VNS), 1):
    #print("n is: {0}".format(n))
    LI_VNS.append((HR_t1VNS[n]-HR_t2VNS[n])/(0.5*(HR_t1VNS[n]+HR_t2VNS[n])))
print(LI_VNS)

```



Great job! Now find the Learning Index of the unstimulated hit rates!




# Exercise 17

```python
# Answer
LI_UN=[]
for n in range(0, len(HR_t1UN), 1):
    #print("n is: {0}".format(n))
    LI_UN.append((HR_t1UN[n]-HR_t2UN[n])/(0.5*(HR_t1UN[n]+HR_t2UN[n])))
print(LI_UN)
```

# Organizing the data to look at it another way


It certainly looks like there's an effect based on the box plot, where vagus nerve stimulation seems to enhance learning about discriminating between two tones. But could there be a more clear way of seeing the relationship between VNS and successful discrimination? Let's convert our Learning Index information into a binary matrix of "shows discrimination" versus "doesn't show discrimination". How can we do that? 
Hint: You can use for loops again!  

# Exercise 19

```python
# Answer
LI_VNS_pos=[]
for n in range(0, len(LI_VNS), 1):
    #print("n is: {0}".format(n))
    if LI_VNS[n]> 0:
        LI_VNS_pos.append(1)
    
print(LI_VNS_pos)

LI_UN_pos=[]
for n in range(0, len(LI_UN), 1):
    #print("n is: {0}".format(n))
    if LI_UN[n]> 0:
        LI_UN_pos.append(1)
    
print(LI_UN_pos)
```

Next, let's convert that information into a percentage. 

# Exercise 20

```python
# Answer
LI_VNS_per=(sum(LI_VNS_pos)/len(LI_VNS))*100
LI_VNS_per
LI_UN_per=(sum(LI_UN_pos)/len(LI_UN))*100
print(LI_VNS_per, LI_UN_per)
```
